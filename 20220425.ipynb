# Auto corelation 이란?

시계열 데이터에서 지연 함수로서 지연된 자기 자신을 가진 신호의 상관관계이다.
즉, 연속된 두 시간 간격 사이의 동일한 변수의 상관 정도를 나타낸다. 변수 값의 지연된 버전이 시계열에서 변수의 원래 버전과 어떻게 관련이 있는지 측정한다.

신호 자기 자신을 시간차를 두고 비교하는 것이다. 

만약 time periodic한 데이터에서 autocorrleation이 높다면, 해당 신호는 유사성을 가져 예측할 수 있다.

# RMS : root mean square 란?



rms값은 평균값과 유사한 성격의 값인데, +,-의 극성과 크기가 변화하고 이런 변화가 주기적으로 반복되는 주기를 갖는다면, 그냥 평균을 구하면 0이 될 수 있다. 

이런 평균값을 제대로 구하기 위해 rms값을 활용한다.

변수(신호의 값) 의 평균의 제곱근을 구함으로써, 실제의 값과 측정된 값의 차이를 보여줄 수 있다. 
(원래의 값을 제곱한 값들의 평균값에 루트를 씌운 값을 계산한다!)

시그널의 사이즈를 강조한다.  이를 통해 진폭이 큰 부분을 강조해줄 수 있다. 

코드로 구현시, 에러값을 구해, 이 알고리즘의 결과로 나오는 에러값이 낮다면 정확도는 높음을 의미한다.
